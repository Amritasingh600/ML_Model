```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("iamsouravbanerjee/house-rent-prediction-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv("/kaggle/input/house-rent-prediction-dataset/House_Rent_Dataset.csv")
```

```python
dt.head()
```

```python
dt.drop(["Posted On","Floor","Area Type","Area Locality","City","Furnishing Status","Tenant Preferred","Point of Contact"],axis=1,inplace=True)
```

```python
X = dt.drop("Rent",axis=1)
```

```python
Y = dt["Rent"]
```

REGULARIZATION

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

BaseLine Model(Linear Regression)

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
print('MSE:',mean_squared_error(Y_test,Y_pred))
print('R2 Score:',r2_score(Y_test,Y_pred))
```

RIDGE REGRESSION

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train,Y_train)
Y_pred = ridge.predict(X_test)
print('MSE:',mean_squared_error(Y_test,Y_pred))
print('R2 Score:',r2_score(Y_test,Y_pred))
```

LASSO

```python
lasso = Lasso(alpha=1.0)
lasso.fit(X_train,Y_train)
Y_pred = lasso.predict(X_test)
print('MSE:',mean_squared_error(Y_test,Y_pred))
print('R2 Score:',r2_score(Y_test,Y_pred))
```

ELASTICNET

```python
elastic = ElasticNet(alpha=1.0,l1_ratio=0.5)
elastic.fit(X_train,Y_train)
Y_pred = elastic.predict(X_test)
print('MSE:',mean_squared_error(Y_test,Y_pred))
print('R2 Score:',r2_score(Y_test,Y_pred))
```

CROSS VALIDATION

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kf = KFold(n_splits=5,shuffle=True,random_state=42)
scores = cross_val_score(model,X,Y,cv=kf,scoring='neg_mean_squared_error')
print('Ridge CV R2 scores:',scores)
print('Lasso CV R2 scores:',scores)
print('ElasticNet CV R2 scores:',scores)
print('Mean CV R2 scores:',scores.mean())
```
